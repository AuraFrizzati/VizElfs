### What is Seaborn?

Seaborn is a Python data visualization library **based on Matplotlib** (which is built on top of numpy). It provides a **high-level interface** for creating attractive and informative statistical graphics.

Seaborn helps you explore and understand your data through:

- **Dataset-Oriented Approach**: **Plotting functions** operate on **dataframes** and **arrays** containing **entire datasets**.
- **Internal Semantic Mapping**: **Automatically maps variables** to **visual attributes**
- **Statistical Aggregation**: Produces informative plots without requiring manual calculations

With its **declarative API**, Seaborn allows you to focus on the meaning of your plots rather than on the technical details of how to draw them.

### Seaborn: An Overview

Seaborn provides a variety of **plotting functions** that share a **common API**. These functions support a wide range of customization options and enable visualization of complex datasets with minimal effort.

- **Plot Types**: From basic scatter and line plots to advanced techniques like regression modeling and kernel density estimation.
    - For Example: The `displot` function can represent **distributions** as **histograms**, **kernel density estimates** or **empirical cumulative distribution functions**.

- **Axes-Level** and **Figure-Level Functions**:
    - **Axes-Level Functions**: Operate like standard Matplotlib functions, allowing **fine-grained control of individual plots**.
    - **Figure-Level Functions**: Automatically manage the creation of figures, enabling features like **faceting** (creating multiple subplots) and customized layouts.

- **Public API for Advanced Applications**: Seaborn provides public classes for **advanced customization**, including direct control over figure layouts.

### Key Features of Seaborn

- **Semantic Mapping:** Automatically maps **dataset variables** to **visual attributes**
- **High-Level Interface**: Simplifies the creation of statistical graphics, focusing on analysis rather than low-level details.
- **Pandas Integration**: Works seamlessly with pandas DataFrames, making it easy to analyze tabular data
- **Attractive Plot Aesthetics**: Built-in **themes** ensure professional-looking plots with minimal effort
- **Statistical Visualizations**: Includes scatter plots, line plots, bar plots, box plots, heatmaps and more
- **Categorical Plotting**: Specialized functions for visualizing grouped or categorized data
- **Multi-Panel Faceting**: Facilitates grids of plots to visualize subsets or categories
- **Flexible Color Mapping**: Includes **color palettes**, **ramps** and **hue mapping** for intuitive visualizations
- **Extensive Customization**: Enables detailed adjustments to plot elements such as titles, labels, legends and more.

### Why Use Distribution Plots?

Distribution plots are essential tools for visualizing the distribution of a dataset, helping analysts and researchers understand the underlying patterns and characteristics of the data. They offer insights into the central tendency, spread and shape of the data, playing a crucial role in data analysis and decision-making.

Below are key reasons to use distribution plots:

- **Identify Data Patterns**: Distribution plots are invaluable for identifying patterns and trends within a dataset. They help visualize how the data is distributedâ€”whether it adheres to a **normal distribution**, **exhibits skewness**, includes **outliers** or displays **multimodal characteristics**.
- **Assess Central Tendency**: These plots provide critical insights into the central tendency of the data, such as the **mean**, **median** or **mode**. Observing the peak or center of the distribution allows us to determine the **most typical** or **representative values**. (if **mean = median = mode**, the data follows a **normal distribution**)
- **Examine Data Spread**: Distribution plots effectively illustrate the **spread** or **variability** of the data. They highlight the **range of values** and the **concentration of data points around the central tendency**. Key spread indicators like the **standard deviation** and **interquartile range** can also be assessed from these plots.
- **Outlier Detection**: Outliersâ€”extreme values that significantly deviate from the rest of the datasetâ€”are easily spotted using distribution plots. These **anomalies** might indicate **errors** in data collection or provide insights into **rare events** and **extreme observations**.
- **Check for Skewness and Symmetry**: **Skewness**, which **describes the asymmetry of a dataset**, can be observed through distribution plots. These plots reveal whether the data is **negatively skewed** (**left-skewed)** or **positively skewed** (**right-skewed**). Understanding skewness helps determine the **data's characteristics** and informs the **selection of appropriate analysis methods**.
- **Compare Distributions**: Distribution plots facilitate the **visual comparison of multiple datasets** or **subgroups**. They help assess differences in distribution shapes, central tendencies and spread across various categories or time intervals.
- **Model Selection**: These plots are useful in **statistical modeling**, as they assist in **model selection** and **assumption validation**. If a statistical model assumes a specific data distribution, distribution plots can verify whether the data meets that assumption or if alternative models need consideration.
- **Communicate Findings**: Distribution plots provide an intuitive and visually appealing way to communicate data characteristics and insights. They offer a **concise summary of the data's distribution**, making it easier to share findings with stakeholders or colleagues.

By leveraging distribution plots, data analysts and researchers gain deeper insights into their data, make informed decisions and effectively present findings.

Different Distribution Plots in Seaborn

Seaborn offers **four main types of distribution plots** for analyzing **univariate** and **bivariate distributions**:

- **Distplot**
- **Jointplot**
- **Pairplot**
- **Rugplot**

Additionally, Seaborn provides **built-in datasets**. For demonstration purposes, we will use the "tips" dataset, which contains information about restaurant bills, tips and related customer demographics.

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
# ignore warnings
filterwarnings('ignore')
```

```python
# load the dataset
df = sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

```python
# set the backgorund style
sns.set_style('whitegrid')
```

#### Displot
`displot` is used for visualising **univariate distributions** through **histograms**

```python
sns.distplot(df['total_bill'], kde = True, color = 'red', bins = 30)
```

**Explanation**:

- **KDE**: **Kernel Density Estimation** adds a **smooth curve** to the histogram
- **Bins**: Number of bins in the histogram, depending on the dataset
- **Color**: Specifies the plot's color

Insight: From the plot, we can infer that most total bills lie between 10 and 20.

#### Jointplot
The `jointplot` visualizes **bivariate** and **univariate distributions** by **combining two plots**.

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

**Explanation**:

- **Kind**: Defines the **visualization type** (**default** is **scatter**; options include hex, reg, kde)
- **X**, **y**: Strings representing **column names** of the dataset
- **Hue**: Adds **categorical separation** between entries in the dataset
- **Palette**: Defines plot colors

Insight: The linear relationship between total bill and tips shows that tips increase as the total bill rises.

#### Pairplot
The `pairplot` displays **pairwise relationships** in a dataset, creating a **grid of Axes**. Each variable is shared across rows and columns for comparisons.

```python
# create pairplot with 'hue' as 'sex'
sns.pairplot(df, hue = 'sex')
```

```python
# create pairplot with 'hue' as 'day'
sns.pairplot(df, hue = 'day')
```

**Insight**: **Pairplots** allow for **detailed exploration** of **relationships** across **multiple variables**.

#### Rugplot
The `rugplot` visualizes **marginal distributions** by adding **ticks** (or **rugs**) along the x and y axes.

This function is intended to **complement other plots** by showing the l**ocation of individual observations** in an unobtrusive way.

https://seaborn.pydata.org/generated/seaborn.rugplot.html

```python
sns.kdeplot(df, x = "tip")
sns.rugplot(df, x = "tip")
```

### Figure Size

You can control the size of your Seaborn plots by specifying the desired **width** and **height** using the `figure()` function from the `matplotlib.pyplot` module. This is particularly useful for adjusting the dimensions of plots to improve readability and presentation.

### Plot Titles and Labels
Adding **titles** and **labels** to your Seaborn plots helps provide **clear descriptions** and **context** for your visualizations.

```python
ax = sns.distplot(df['total_bill'], kde = True, color = 'red', bins = 30)
ax.set(
    xlabel = "Total Bill", 
    ylabel = "Frequency", 
    title = "Histogram Total bill")
```

### Fonts and Font Sizes

Seaborn uses Matplotlib for **font settings**, allowing you to customize the **fonts** and **font sizes** in your plots. You can adjust these **settings globally** using Matplotlib's `rcParams`.

```python
# Set the font family to Arial
plt.rcParams["font.family"] = "Arial"
# Set the font size to 12
plt.rcParams["font.size"] = 12

ax = sns.distplot(df['total_bill'], kde = True, color = 'red', bins = 30)
ax.set(
    xlabel = "Total Bill", 
    ylabel = "Frequency", 
    title = "Histogram Total bill")
```

### Grid Appearance

Seaborn plots typically include **grid lines** by default, enhancing the **readability** of the visualizations. You can further customize the grid's appearance using the `grid()` function from Matplotlib's pyplot module.

```python
sns.set_theme(style="whitegrid")
ax = sns.distplot(df['total_bill'], kde = True, color = 'red', bins = 30)
ax.set(
    xlabel = "Total Bill", 
    ylabel = "Frequency", 
    title = "Histogram Total bill")
ax.grid(color ="black", linestyle="--", linewidth = 0.5)
```

### Axes Style

The style of the axes in Seaborn plots can be customized to enhance their appearance. Specifically, the **spines** attribute of **axes objects** allows for precise control over the **borders** (**spines**) of the plot.

```python
sns.set_theme(style="whitegrid")
ax = sns.distplot(df['total_bill'], kde = True, color = 'red', bins = 30)
ax.set(
    xlabel = "Total Bill", 
    ylabel = "Frequency", 
    title = "Histogram Total bill")

ax.spines["left"].set_linewidth(3)    # Set the width of the left spine
ax.spines["bottom"].set_linewidth(3)  # Set the width of the bottom spine
```

### Pie Chart

Seaborn **does not have a built-in function** specifically for creating pie charts as it **primarily focuses on statistical visualization**. However, you can use `Matplotlib` to create pie charts. Below are examples of how to generate pie charts:

```python
# Sample data
sizes = [30, 25, 15, 10, 20]
labels = ['A', 'B', 'C', 'D', 'E']

# Define Seaborn color palette
palette_color = sns.color_palette('bright')

# Create the pie chart
plt.pie(sizes, labels=labels, colors=palette_color, autopct='%1.1f%%')

# Set the aspect ratio to make it a circle
#plt.axis('equal')

# Set the chart title
plt.title('Pie Chart')

# Show the plot
plt.show()
```

### Bar Chart
Bar plots are a popular choice for visualizing and comparing **categorical data** due to their versatility and clarity.

Below are the key reasons why bar plots are so effective:
- **Comparison of Categories**: Bar plots clearly represent **values** or **frequencies** for different categories, enabling easy comparison and identification of patterns or differences.
- **Categorical Data Visualization**: They are particularly suited for representing categorical data, where each category corresponds to a distinct group or label, effectively showcasing qualitative or discrete data.
- **Distribution of Data**: Bar plots illustrate the distribution or frequency of data within each category. The height of each bar reflects the magnitude or frequency of the variable.
- **Trend Analysis**: These plots can reveal how categorical data changes **over time** or **under different conditions**. Bar plots created for multiple time periods or conditions help identify trends and variations.
- **Comparison of Subcategories**: By incorporating subcategories or grouping variables, bar plots facilitate comparisons between subgroups within each category, offering deeper insights into the data.
- **Communicating Insights**: Intuitive and visually appealing, bar plots are excellent tools for communicating findings to a wide audience. They simplify complex data, making it easier to understand categorical relationships.
- **Exploratory Data Analysis**: Bar plots are powerful exploratory tools for gaining initial insights into data. They help identify **outliers**, **anomalies** and **relationships** between **categorical variables**.
- **Decision Making**: These plots support data-driven decision-making by visualizing important categories, trends and patterns, aiding in informed choices.

Seaborn provides a user-friendly and high-level interface for creating bar charts through its `barplot` function. This function offers capabilities for statistical estimation and extensive customization, making it a valuable tool for effective data visualization.

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
subcategories = ['Subcategory 1', 'Subcategory 2', 'Subcategory 1', 'Subcategory 2']
values = [25, 40, 30, 35]

# Create the bar chart
sns.barplot(x = categories, y = values, palette = "bright")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
```

Additionally, you can enhance the appearance of the bar chart by leveraging various parameters available in the barplot function. For instance, the `hue` parameter enables **grouping bars based on a categorical variable** and the `palette` parameter allows you to specify a **color palette** for the chart.

```python
# Sample data
categories = ['Category A', 'Category A', 'Category B', 'Category B']
subcategories = ['Subcategory 1', 'Subcategory 2', 'Subcategory 1', 'Subcategory 2']
# Create the bar chart
sns.barplot(
    x = categories, 
    y = values, 
    palette = "bright", 
    hue = subcategories
    , legend = "auto"
    # ,dodge = False
    # ,gap = True
    )
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Hue")
```

```python
# Create grouped bar chart plot
sns.catplot(
    x = categories, 
    y = values, 
    palette = "bright", 
    hue = subcategories
    ,legend = "auto"
    ,kind = "bar"
    )
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Hue")
```
